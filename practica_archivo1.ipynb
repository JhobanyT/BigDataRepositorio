```python
url = ("https://raw.githubusercontent.com/JhobanyT/BigDataRepositorio/main/archivo_1.csv")
df = pd.read_csv(url)
print(df.shape)
df.head()
```

```python
df.columns.values
```

Obtenemos la suma total de ingresos

```python
suma_ingresos = df['ingresos'].sum()
print("La suma de los inresos es: ", suma_ingresos)
```

Promedio de edades

```python
promedio_edad = df['edad'].mean()
print("El promedio de la edad de las persona es:", int(promedio_edad))
```

Obtenemos las personas que trabajan en arequipa

```python
personas_arequipa = df[df['ciudad'] == "Arequipa"].sum
personas_arequipa.iloc[:, [0,2]]
```

Obtenemos Los ingresos por ciudad

```python
ingreso_ciudad = df.groupby('ciudad')['ingresos'].mean().astype(int)
ingreso_ciudad
```

La persona más joven

```python
# persona mas joven
persona_joven = df.nsmallest(1, 'edad')
print(f"La persona mas joven es: {persona_joven['nombre']}")
```

La persona más mayor

```python
# persona mas mayor
persona_mayor = df.nlargest(1, 'edad')
print(f"La persona mas mayor es: {persona_mayor['nombre']}")
```

Trabajadores por ciudad

```python
# primera opcion
trabajadores_ciudades = df.groupby('ciudad')['nombre'].count()
trabajadores_ciudades
```

```python
# segunda opcion
personas_ciudades = df['ciudad'].value_counts()
personas_ciudades
```

Orden de personas con mayor edad

```python
# Primera opcion
edad_orden = df.nlargest(len(df["edad"]),'edad')
edad_orden
```

```python
# Segunda opción
edad_orden = df.sort_values("edad", ascending=False)
edad_orden
```

TAREA

```python
ingresos_mayores_30 = df[df["edad"] > 30]["ingresos"].mean()
print(f"El promedio de ingresos de las personas mayores a 30 años es: {int(ingresos_mayores_30)}")
```

Promedio de ingresos por ciudades

```python
ingresos_ciudades = df.groupby('ciudad')['ingresos'].sum()
ingresos_ciudades
```

Personas con el ingreso mas alto

```python
persona_mayor_ingreso = df.nlargest(1, 'ingresos')
print(f"La persona con el mayor ingreso es: {persona_mayor_ingreso['nombre']}")
```
